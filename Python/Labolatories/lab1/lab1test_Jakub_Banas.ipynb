Utwórz string `s1` będący powtórzeniem Twojego imienia i nazwiska tyle razy ile wynosi reszta z dzielenia wartości Twojego numeru indeksu przez `99` (jeśli reszta wynosi `0`, to weź resztę z dzielenia przez `999`). Imię i Nazwwisko ma być rozdzielone podkreślnikiem `_` 

przykład:
```
nr_indeksu, imie, nazwisko = 102, 'Błażej', 'Kuś'
s1: Błażej_Kuś_Błażej_Kuś_Błażej_Kuś
```

**znajdź:**
`w1`: dlugość `s1`

Utwórz string `s2` będący kopią `s1` bez pierwszego i ostatniego znaku i zamień w nim wszystkie znaki różne od liter `aeiouyAEIOUY` na literę x

`w2`: to suma kodów (Unicode) wszystkich znaków w `s2`

`w3`: to liczba wystąpień litery `a` w `s1`

`w4`: porównaj (leksykograficznie) stringi  `s1` i `s2` i zwróć Unicode ostatniego znaku mniejszego stringa

`w5`: dodaj wynik porównania `w3` z `w4` do pierwszej cyfry Twojego numeru indeksu
(uwaga: w tym przypadku to jest porównanie na zasadzie obie wartości są równe albo nie, a nie tak jak dla w4 gdzie jest porównujemy który string jest mniejszy a który większy)  

````python
nr_indeksu, imie, nazwisko, s1= '151843', 'Jakub', 'Banaś', ''

# mod = nr_indeksu % 99 #dzielenie modulo
mod = int(nr_indeksu) % 99 if int(nr_indeksu) % 99 else int(nr_indeksu) % 999
print(mod)

#dopisanie _ miedzy imieniem i nazwiskiem
for i in range (1, mod):
    s1 += imie + '_' + nazwisko + '_'
print(s1)

# dlugosc s1
w1 = len(s1)
print(f"\ndlugosc s1 = {w1}\n")

#zamiana spolglosek na x
nowy = ""
for i in s1:
    if i not in "aeiouyAEIOUY":
        nowy += "x"
    else:
        nowy += i
s2 = nowy      # przypisanie znakow nowy do s2
s2 = s2[1:-1]  #usuniecie pierwszego i ostatniego znaku
print(s2)

#dlugosc s2
w2 = len(s2)
print(f"\ndlugosc w2 = {w2}\n")

#ilosc wystapien litery a
w3 = 0
for i in range(w2):
    if s2[i] == 'a':
        w3 += 1
print(f"\nilosc wystapien litery a = {w3}\n")

#ostatni znak krotszego stringa
print("ostatni znak mniejszego stringa: ")
if w4 == s1 < s2:
    print(f"s1[-1] = {s1[-1]}")
else:
    print(f"s2[-1] = {s2[-1]}")


#porownanie w3 z w4 i dopisanie wartosci porownania na poczatku indeksu
w5 = (w4 == w3)
if w5:
    print("w3 == w4, wynik 1\n")
    nr_indeksu = "1" + nr_indeksu
else:
    print("w3 != w4, wynik 0\n")
    nr_indeksu = "0" + nr_indeksu

print(f"nr_indeksu = {nr_indeksu}")
````
