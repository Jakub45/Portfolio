## Zadanie
z d1 usuń elementy których klucze nie zawierają żadnej cyfry z Twojego nr indeksu - oprócz `'0'`
a ich wartości nie zawierają żadnej litery z Twojego imienia ani nazwiska
i utwórz z nich nowy słownik d2

utwórz zbiór z1 zawierający wszystkie cyfry wystąpujące w kluczach d2
oraz litery występujące w wartościach elementów pozostałych w d1

zwróć w1 równe: liczebność zbioru z1 pomnożona przez nr_indeksu 

Uwaga!!! 
cyfry pisane zarówno jako znaki jak i liczbowo potraktuj jako znaki ASCII:
zarówno 5 jak i '5' to cyfra '5'

```python
d1 = {
    '224': 'weer',
    '213': 'asgfd',
    '234': 'asdf',
    '233': 'ads',
    (0, 1, 2): 'jeden',
    (1, 2, 3): 'dwa',
    (1, 2, 4): ['a', 'b', 'p'],
    tuple([1, 2, 3, '5']): set(['a', 'b', 'c', 'e', 's', 'e', 'c', 'a'])
}

nr_indeksu, imie, nazwisko = 151843, 'Jakub', 'Banaś'

indeks_str = set(str(nr_indeksu))
imie_set = set(imie.lower())
nazwisko_set = set(nazwisko.lower())

# --- 1. filtrujemy d1 -> d2
d2 = {}

for k, v in list(d1.items()):

    k_str = set(str(k))

    # warunek 1: klucz NIE zawiera cyfr z indeksu (poza '0')
    if any(ch in indeks_str and ch != '0' for ch in k_str):
        continue

    # warunek 2: wartości NIE zawierają liter imienia/nazwiska
    v_str = set(str(v).lower())

    if v_str & imie_set or v_str & nazwisko_set:
        continue

    d2[k] = v

# --- 2. zbiór z1
z1 = set()

# cyfry z kluczy d2
for k in d2.keys():
    for ch in str(k):
        if ch.isdigit():
            z1.add(ch)

# litery z wartości pozostałych w d1 (czyli NIE w d2)
for k, v in d1.items():
    if k not in d2:
        for ch in str(v).lower():
            if ch.isalpha():
                z1.add(ch)

# --- 3. wynik
w1 = len(z1) * nr_indeksu

print("d2:", d2)
print("z1:", z1)
print("w1:", w1)
```
